# Top 10 highest-rated iphone on Flipcart in india?


```python
data.head()
```

```python
highest_rated = data.sort_values(by = ["Star Rating"],ascending = False)
```

```python
highest_rated = highest_rated.head(10)
```

```python
print(highest_rated['Product Name'])
```

```python
print(highest_rated)
```

#How many ratings do the highest rated iphones on flipcart have?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated,x=label,y= counts)
figure.show()
```

#which iphone has the highest number of reviews on flipcart?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Reviews']
figure = px.bar(highest_rated,x=label,y= counts,title = "number of reviews of high rated iphone")
figure.show()
```

#What is the relationship between the sale price of iphones and the number of ratings on flipcart?

```python
data.head()
```

```python
figure = px.scatter(data_frame=data,x="Number Of Ratings",y = "Sale Price",size = "Discount Percentage",trendline ="ols")
figure.show()
```

#What is the relationship between the discount rate and the number of ratings on flipcart?

```python
figure = px.scatter(data_frame=data,x="Number Of Ratings",y = "Discount Percentage",size = "Sale Price",trendline ="ols")
figure.show()
```

#Most expensive and least expensive

```python
data.head()
```

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

#Display the results
print("Most Expensive Iphone:")
print(most_expensive)
print("\nLeast Expensive Iphone:")
print(least_expensive)
```
